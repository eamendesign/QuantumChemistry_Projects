```python
from ase.build import molecule
from ase.calculators.emt import EMT
from ase.optimize import BFGS
from ase.vibrations import Vibrations

# 1. 建立 H2O 分子
atoms = molecule("H2O")

# 2. 指定计算器
# EMT 适合教学演示；真实频率建议换成 DFT 计算器
atoms.calc = EMT()

# 3. 先做几何优化
opt = BFGS(atoms, trajectory="h2o_opt.traj", logfile="opt.log")
opt.run(fmax=0.01)

# 4. 振动分析
vib = Vibrations(atoms, name="h2o_vib")
vib.run()

# 5. 输出频率总结
vib.summary()

# 6. 可选：写出振动模式文件
# vib.write_mode()

# 7. 结束后清理位移计算生成的临时文件
vib.clean()
```
